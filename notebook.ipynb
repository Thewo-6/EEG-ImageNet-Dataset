```python
import torch

path1 = r"D:\DATASETS\EEG-ImageNet\EEG-ImageNet_1.pth"
path2 = r"D:\DATASETS\EEG-ImageNet\EEG-ImageNet_2.pth"
out_path = r"D:\DATASETS\EEG-ImageNet\EEG-ImageNet.pth"

# Load both files
data1 = torch.load(path1, map_location="cpu")
data2 = torch.load(path2, map_location="cpu")

# Merge the two dictionaries
merged = {}
for key in data1.keys():
    v1 = data1[key]
    v2 = data2[key]

    if isinstance(v1, list) and isinstance(v2, list):
        merged[key] = v1 + v2       # simple list concatenation
    else:
        # fallback, just keep both if ever needed
        merged[key] = [v1, v2]

# Quick sanity check
for k in merged:
    if isinstance(merged[k], list):
        print(k, "length:", len(merged[k]))

# Save merged file
torch.save(merged, out_path)
print("Saved merged dataset to:", out_path)
```
